```python
pattern='\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages=re.split(pattern,data)[1:]
```

```python
messages
```

```python
dates=re.findall(pattern,data)
```

```python
dates
```

```python
df = pd.DataFrame({'user_messages':messages,'message_dates':dates})
```

```python
df.head()
```

```python
df['message_dates'] = pd.to_datetime(df['message_dates'], format='%d/%m/%y, %H:%M - ')
```

```python
df.rename(columns={'message_dates':'dates'},inplace = True)
```

```python
len(df)
```

```python
df.shape
```

```python
users=[]
messages=[]
for message in df['user_messages']:
    entry=re.split('([\w\W]+?):\s',message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('Group Notifications')
        messages.append(entry[0])
```

```python
df['users']=users
df['message']=messages
df.drop(columns='user_messages',inplace=True)
```

```python
period = []
for hour in df['hour']:
    if hour == 23:
        period.append(str(hour) + "-" + str('00'))
    elif hour == 0:
        period.append(str('00') + "-" + str(hour + 1))
    else:
        period.append(str(hour) + "-" + str(hour + 1))

df['period'] = period
```
